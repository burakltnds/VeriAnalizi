## Model EÄitimine BaÅla

```python
y = df["price"].values
x=df.drop("price",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test , y_train , y_test = train_test_split(x,y,test_size=0.3 ,random_state=10)
```

```python
# kontrol
len(x_test)
len(x_train)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape #5 katmandan baÅlatsak iyi olur
```

```python
model = Sequential()
model.add(Dense(8,activation="relu"))
model.add(Dense(8,activation="relu"))
model.add(Dense(8,activation="relu"))
model.add(Dense(8,activation="relu"))
model.add(Dense(8,activation="relu")) # 12 nÃ¶ronlu 4 katmanÄ±mÄ±z oldu aktvasyon fonksiyonumuz relu

model.add(Dense(1)) # boyut

model.compile(optimizer ="adam" ,loss="mse") #optimizer adam fonksÅyonu loss fonksiyonuda mse
```

```python
model.fit(x=x_train , y=y_train , validation_data=(x_test , y_test) ,batch_size=250,epochs=350)
#eÄitim baÅlar batchsize=kaÃ§arlÄ± inceleyyeceÄi epochs Ã¼zerinden kaÃ§ kez geÃ§eceÄi 
#validation data test verileri
```

## SonuÃ§larÄ± DeÄerlendirme

```python
# KayÄ±p verisi 
kayip = pd.DataFrame(model.history.history)
kayip.head(10)
```

```python
kayip.plot()
```

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error
tahminDizisi = model.predict(x_test)
```

```python
mean_absolute_error(y_test , tahminDizisi)
```
